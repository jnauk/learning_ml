```
#Exercise 4
!wget -O mnist.csv -nc "https://drive.google.com/uc?export=download&id=1_IsyUAT5aehyBOaHuwXI-qLrAmmbQdZF"

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 

df = pd.read_csv('mnist.csv')
M = df.to_numpy()
im = M[3][1:]
im = im.reshape(28, 28)

plt.imshow(im)
plt.show()

size = 28
for layer in range(size//2):
  layerLimit = size - layer -1;
  fwd = layer
  bwd = layerLimit;
  while fwd < layerLimit:
    tmp = im[fwd][layer]
    im[fwd][layer] = im[layerLimit][fwd]
    im[layerLimit][fwd] = im[bwd][layerLimit]
    im[bwd][layerLimit] = im[layer][bwd]
    im[layer][bwd] = tmp
    fwd += 1
    bwd -= 1
plt.imshow(im)
plt.show()

im = np.rot90(im, axes=(1,0))
plt.imshow(im)
plt.show()
```

```
#Exercise 5

import numpy as np

def is_symmetric(matrix):
  result = True
  shape = matrix.shape
  if shape[0] != shape[1]:
    result = False
  else:
    size = shape[0]
    for layer in range(size//2):
      layerLimit = size - layer -1
      for fwd in range(layer, layerLimit+1):
        if matrix[layer][fwd] != matrix[fwd][layer] or matrix[layerLimit][fwd] != matrix[fwd][layerLimit]:
          result = False
          break
      if not result:
        break
  return result

M = np.array([[1, 2, 3],
             [2, 2, 3],
             [3, 3, 4]])
print("OK" if is_symmetric(M) else 'ERROR')
print("OK" if np.allclose(M, M.T) else 'ERROR')

M = np.array([[1, 2, 3],
             [2, 2, 3]])
print("OK" if not is_symmetric(M) else 'ERROR')
#print("OK" if not np.allclose(M, M.T) else 'ERROR')

M = np.array([[1, 3, 3],
             [2, 2, 3],
             [3, 3, 4]])
print("OK" if not is_symmetric(M) else 'ERROR')
print("OK" if not np.allclose(M, M.T) else 'ERROR')
```

```
#Exercise 8
import numpy as np
from matplotlib import pyplot as plt 
import pandas as pd

size = 100
ang_width = 0.2
ang_max = np.pi*0.5
r_max = 10


def add_spiral(data, angle, color):
  ang = np.random.uniform(0,ang_max, size)
  r = ang * r_max

  c = np.empty(size, dtype='object')
  c.fill(color)
  data2 = np.array([np.sqrt(r) * np.cos(ang+angle), np.sqrt(r) * np.sin(ang+angle)]).T
  data2 = data2 + ang_width*np.random.randn(size,2) #np.random.uniform(0,np.pi, size)
  data2 = np.c_[data2, c]
  if data is None:
    data = data2
  else :    
    data = np.concatenate((data, data2))

  return data

data = None

data = add_spiral(data, 2*np.pi*(1/3), 'red')
data = add_spiral(data, 2*np.pi*(2/3), 'red')
data = add_spiral(data, 2*np.pi*(3/3), 'red')
data = add_spiral(data, 2*np.pi*(1/3)+np.pi, 'blue')
data = add_spiral(data, 2*np.pi*(2/3)+np.pi, 'blue')
data = add_spiral(data, 2*np.pi*(3/3)+np.pi, 'blue')

#plt.scatter(x, y)
plt.scatter(data[:,0], data[:,1], c=data[:,2])
plt.axis('equal')
plt.show()

#Exercise 9
df = pd.DataFrame(data);
df.columns = ['x1', 'x2', 'y']
df.to_csv('data.csv', index=False)
```
